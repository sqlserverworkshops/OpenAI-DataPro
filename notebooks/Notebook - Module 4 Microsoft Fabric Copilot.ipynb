```python
df = spark.sql("SELECT * FROM baseball_Lakehouse.movieInfo")
display(df)
```

```python
%%code
Remove the dollar sign, commas, and white space from the AveragePerTheater column.  Place that value in a new column titled AvgPerTheaterValue and append it to the previous dataframe.  The data type for AveragePerTheater should remain a string.  Display the top 10 rows. Display the top 10 rows. Do this same set of actions for the TheaterCount column and place the results in a new column named TotalTheaters, keep the data type a string. 
```

```python
import pyspark.sql.functions as F

df = df.withColumn("AvgPerTheaterValue", F.regexp_replace(F.col("AveragePerTheater"), "[$, ]", ""))
df = df.withColumn("TotalTheaters", F.regexp_replace(F.col("TheaterCount"), "[$, ]", ""))
df.show(10)
```

```python
%%code
In the df dataframe convert the AvgPerTheaterValue to a numeric value and append it to the df dataframe retaining all columns replacing the AvgPerTheaterValue column. Do this same set of actions for the TotalTheaters column.
```

```python
df = df.withColumn("AvgPerTheaterValue", F.col("AvgPerTheaterValue").cast("float"))
df = df.withColumn("TotalTheaters", F.col("TotalTheaters").cast("float"))
df.show(10)
```

```python
%%code
Summarize the TotalTheaters column by MovieTitle in the df dataframe.  Create a bar chart of the top 10 films by max TotalTheaters, display MovieTitle as part of the chart.
```

```python
import matplotlib.pyplot as plt

top_10_films = df.groupBy("MovieTitle").agg(F.max("TotalTheaters").alias("MaxTotalTheaters")).orderBy(F.desc("MaxTotalTheaters")).limit(10)

film_titles = [row["MovieTitle"] for row in top_10_films.collect()]
max_total_theaters = [row["MaxTotalTheaters"] for row in top_10_films.collect()]

plt.bar(film_titles, max_total_theaters)
plt.xlabel("Movie Title")
plt.ylabel("Max Total Theaters")
plt.title("Top 10 Films by Max Total Theaters")
plt.xticks(rotation=90)
plt.show()
```

```python
%%code
Summarize the AveragePerTheaterValue column by MovieTitle in the df dataframe.  Create a bar chart of the top 10 films by max AveragePerTheaterValue, display MovieTitle as part of the chart.
```

```python
import matplotlib.pyplot as plt

top_10_films_avg = df.groupBy("MovieTitle").agg(F.max("AvgPerTheaterValue").alias("MaxAvgPerTheaterValue")).orderBy(F.desc("MaxAvgPerTheaterValue")).limit(10)

film_titles_avg = [row["MovieTitle"] for row in top_10_films_avg.collect()]
max_avg_per_theater_value = [row["MaxAvgPerTheaterValue"] for row in top_10_films_avg.collect()]

plt.bar(film_titles_avg, max_avg_per_theater_value)
plt.xlabel("Movie Title")
plt.ylabel("Max Average Per Theater Value")
plt.title("Top 10 Films by Max Average Per Theater Value")
plt.xticks(rotation=90)
plt.show()
```
